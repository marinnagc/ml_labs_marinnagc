## Número de linhas para acomodar figuras

Para ajudar a arrumar os *plots*, vou fazer um truque de aritmética aqui. Seja $n_{\text{graphs}}$ o número de figuras que temos que plotar, e $n_{\text{cols}}$ o número de colunas que queremos ter. Qual dever ser o número de linhas $n_{\text{rows}}$?

> ***Proposição***
>
> O número de linhas $n_{\text{rows}}$ deve ser
>
> $$
> n_{\text{rows}} = \left \lfloor \frac{n_{\text{graphs}} - 1}{n_{\text{cols}}} \right \rfloor + 1
> $$

***Prova***:

> ***Nota***: Divisão inteira.
> 
> Em primeiro lugar, os símbolos misteriosos $\lfloor$ e $\rfloor$ são a *função piso*. 
>
> **Função piso**: retorna o maior inteiro que seja menor ou igual ao valor real no argumento da função. Ou seja: "arredonda para a direção negativa". Por exemplo, 
>
> - $\lfloor 3.14 \rfloor = 3$
> - $\lfloor 15 \rfloor = 15$
>
> Para valores negativos a função opera do mesmo jeito, mas cuidado: ela arredonda para a direção negativa, e não para "perto do zero"!
>
> - $\lfloor -1.67 \rfloor = -2$
> - $\lfloor -15 \rfloor = -15$
>
> Portanto, para valores $a$ e $b$ inteiros, temos:
> 
> $$
> \left \lfloor \frac{a}{b} \right \rfloor = \text{Resultado da divisão inteira de } a \text{ por } b
> $$
> 
> Em Python, para $a$ e $b$ inteiros, basta fazer `a // b` para obter o valor de $\lfloor a / b \rfloor$.

Vamos considerar dois casos:

- *Quando $n_{\text{graphs}}$ é um múltiplo de $n_{\text{cols}}$*

Neste caso a resposta desejada para $n_{\text{rows}}$ seria $n_{\text{rows}} = n_{\text{graphs}} / n_{\text{cols}}$ diretamente, e todas as linhas estarão cheias.

Ao remover $1$ do denominador e aplicar a função "piso", obtemos um valor que é exatamente $1$ a menos que o valor desejado, pois agora $(n_{\text{graphs}} - 1)$ **não é mais divisível por $n_{\text{cols}}$**. 

$$
(n_{\text{cols}} \mid n_{\text{graphs}})
\Rightarrow
\left \lfloor \frac{n_{\text{graphs}} - 1}{n_{\text{cols}}} \right \rfloor = 
\frac{n_{\text{graphs}}}{n_{\text{cols}}} - 1
$$

(Nota: "$n_{\text{cols}} \mid n_{\text{graphs}}$" lê-se "$n_{\text{cols}}$ divide $n_{\text{graphs}}$" ou "$n_{\text{graphs}}$ é divisível por $n_{\text{cols}}$")

Quando somamos $1$ ao resultado da divisão, recuperamos o resultado desejado:

$$
\left \lfloor \frac{n_{\text{graphs}} - 1}{n_{\text{cols}}} \right \rfloor + 1 =
\left( \frac{n_{\text{graphs}}}{n_{\text{cols}}} - 1 \right) + 1=
\frac{n_{\text{graphs}}}{n_{\text{cols}}}
$$

Logo,

$$
n_{\text{rows}} = \left \lfloor \frac{n_{\text{graphs}} - 1}{n_{\text{cols}}} \right \rfloor + 1
$$

- *Quando $n_{\text{graphs}}$ não é um múltiplo de $n_{\text{cols}}$*

Neste caso a resposta desejada é a divisão inteira de $n_{\text{graphs}}$ por $n_{\text{cols}}$ (que é o número de linhas "cheias") mais $1$ (para acomodar o restante das figuras).

A divisão inteira de $n_{\text{graphs}}$ por $n_{\text{cols}}$ é

$$
\left \lfloor \frac{n_{\text{graphs}}}{n_{\text{cols}}} \right \rfloor
$$

Como "$n_{\text{cols}} \nmid n_{\text{graphs}}$" (lê-se "$n_{\text{cols}}$ não divide $n_{\text{graphs}}$" ou "$n_{\text{graphs}}$ não é divisível por $n_{\text{cols}}$"), então tem pelo menos $1$ de "gordurinha" em $n_{\text{graphs}}$, e portanto

$$
(n_{\text{cols}} \nmid n_{\text{graphs}})
\Rightarrow
\left \lfloor \frac{n_{\text{graphs}} - 1}{n_{\text{cols}}} \right \rfloor =
\left \lfloor \frac{n_{\text{graphs}}}{n_{\text{cols}}} \right \rfloor
$$

que é o valor de linhas cheias desejado. Basta somar $1$ para ter o número total de linhas necessárias para acomodar todos as $n_{\text{graphs}}$ figuras!

Então

$$
n_{\text{rows}} = \left \lfloor \frac{n_{\text{graphs}} - 1}{n_{\text{cols}}} \right \rfloor + 1
$$

Conclusão, em ambos os casos, a fórmula

$$
n_{\text{rows}} = \left \lfloor \frac{n_{\text{graphs}} - 1}{n_{\text{cols}}} \right \rfloor + 1
$$

resolve nosso problema de calcular o número de linhas para as figuras.
